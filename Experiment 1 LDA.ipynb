```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn 
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn import datasets

iris = datasets.load_iris()
```

```python
df = pd.read_csv("./Datasets/Iris.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
x = df.iloc[:,1:5].values
y = df.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
# Preprocessing the data by dividing into test and training datasets

sc = StandardScaler() # centering all the data points around the mean value 
x = sc.fit_transform(x)
le = LabelEncoder() # assigns a numerical value each class of observations 0,1 and 2 for Iris Dataset
y = le.fit_transform(y) 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x
```

```python
lda = LinearDiscriminantAnalysis(n_components = 2)
x_train_transformed = lda.fit_transform(x_train,y_train)
x_test_transformed = lda.transform(x_test)
```

```python
x_train_transformed
```

```python
y_train
```

```python
x_test_transformed
```

```python
plt.scatter(x_train_transformed[:,0],x_train_transformed[:,1], c=y_train,cmap='rainbow', alpha=0.9, edgecolors = 'b')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('LDA Transformed Data (Training Set)')
plt.show()
```

```python
plt.scatter(x_test_transformed [:, 0], x_test_transformed [:, 1], c=y_test,cmap='rainbow',
alpha=0.7,edgecolors='b')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('LDA Transformed Data (Testing Set)')
plt.show()
```

```python
accuracy_train=lda.score(x_train, y_train)
print("Training Accuracy:", accuracy_train)
```

```python
accuracy_test = lda.score(x_test, y_test)
print("Testing Accuracy:", accuracy_test)
```
