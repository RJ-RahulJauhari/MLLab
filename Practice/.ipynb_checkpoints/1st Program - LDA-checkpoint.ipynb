# Dimensionality Reduction using LDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
sns.pairplot(df, hue = 'species')
```

```python
df.describe()
```

```python
df.iloc[:,0:len(df.columns)-1].corr()
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.sample(5)
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

## Initializing LDA

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33)
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
train_transformed = lda.fit_transform(x_train,y_train)
```

```python
test_transformed = lda.transform(x_test)
```

```python
train_transformed
```

```python
test_transformed
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = train_transformed[:,0], y = train_transformed[:,1], hue = y_train, legend = 'full')
plt.xlabel('Linear Component 1')
plt.ylabel('Linear Component 2')
plt.title("Training Component Analysis")
plt.show()
```

```python
print(f'{lda.score(x_train,y_train)*100}% accuracy')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = test_transformed[:,0], y = test_transformed[:,1], hue = y_test, legend = 'full')
plt.xlabel('Linear Component 1')
plt.ylabel('Linear Component 2')
plt.title("Testing Component Analysis")
plt.show()
```

```python
print(f'{lda.score(x_test,y_test)*100}% accuracy')
```
