```python
import pandas as pd    #read csv files
import matplotlib.pyplot as plt 
import sklearn #preprocessing functions
from sklearn.preprocessing import StandardScaler, LabelEncoder 
from sklearn.model_selection import train_test_split 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
```

```python
# read dataset from URL
url = "./Datasets/Iris.csv"
dataset = pd.read_csv(url)
```

```python
print(dataset)
```

```python
dataset.shape
```

```python
# divide the dataset into class and target variable
X = dataset.iloc[:, 1:5].values
y = dataset.iloc[:, -1].values #iloc extracts
```

```python
X
```

```python
# Preprocess the dataset and divide into train and test
sc = StandardScaler() #mean becomes 0 and std becomes 1
X = sc.fit_transform(X)
le = LabelEncoder()
y = le.fit_transform(y) #it assigns 0 1 and 2 for each class
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # test_size=20% remaining train
```

```python
X
```

```python
# apply Linear Discriminant Analysis
lda = LinearDiscriminantAnalysis(n_components=2) # 4 features will be reduced to 2 # min(no.of features, no.of classes(3)-1)
X_train_Transformed = lda.fit_transform(X_train, y_train) 
# Applying the trained LDA model to the testing data
X_test_Transformed = lda.transform(X_test) 
```

```python
X_train_Transformed
```

```python
y_train
```

```python
X_test_Transformed
```

```python
# plot the scatterplot
plt.scatter(X_train_Transformed[:,0],X_train_Transformed[:,1],
            c=y_train,cmap='rainbow',
alpha=0.9,edgecolors='b')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('LDA Transformed Data (Training Set)')
plt.show()
```

```python
# Plotting the transformed data for the testing set
plt.scatter(X_test_Transformed [:, 0], X_test_Transformed [:, 1], c=y_test,cmap='rainbow',
alpha=0.7,edgecolors='b')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('LDA Transformed Data (Testing Set)')
plt.show()
```

```python
accuracy_train=lda.score(X_train, y_train)
print("Training Accuracy:", accuracy_train)
```

```python
# Computing the accuracy of classification on the testing data
accuracy_test = lda.score(X_test, y_test)
print("Testing Accuracy:", accuracy_test)
```
